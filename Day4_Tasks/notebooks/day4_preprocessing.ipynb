```python
import pandas as pd
```

```python
data = {
    'Age': [21, None, 23, 24],
    'Score': [80, 75, None, 95]
}
df = pd.DataFrame(data)
```

```python
# Drop missing values
df_drop = df.dropna()
```

```python
# Fill missing values
df_fill = df.fillna(df.mean())
```

```python
print("Dropped:\n", df_drop)
print("\nFilled:\n", df_fill)
```

### Why is it better to fill a missing age with the mean rather than just putting 0?

##### - 0 is unrealistic (no person has age 0 in this dataset)
##### - It distorts the data distribution
##### - Mean keeps the dataset balanced and statistically accurate
##### - Helps the model learn better patterns
