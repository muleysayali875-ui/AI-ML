```python
from textblob import TextBlob
```

```python
sample_bios = [
    "I absolutely love attending coding workshops and networking events!",
    "This event was poorly managed and very disappointing.",
    "Had an average experience, nothing too exciting.",
    "Amazing organizers and excellent event planning!",
    "Terrible communication and boring sessions."
]
```

```python
for bio in sample_bios:
    analysis = TextBlob(bio)
    
    print("Bio:", bio)
    print("Polarity:", analysis.sentiment.polarity)
    print("Subjectivity:", analysis.sentiment.subjectivity)
    print("-" * 60)
```

```python
results = []

for bio in sample_bios:
    analysis = TextBlob(bio)
    results.append({
        "Bio": bio,
        "Polarity": analysis.sentiment.polarity,
        "Subjectivity": analysis.sentiment.subjectivity
    })

import pandas as pd
df = pd.DataFrame(results)
df
```

### Documented the difference between "Stemming" and "Lemmatization".
#### Stemming:
- Stemming cuts off prefixes or suffixes to reduce a word to its root, which may not always be a valid word.
- Example: studies → studi

#### Lemmatization:
- Lemmatization converts a word to its proper dictionary base form.
- Example: studies → study
