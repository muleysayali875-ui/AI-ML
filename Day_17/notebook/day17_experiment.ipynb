# DAY 17: HYPERPARAMETER TUNING

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer

# 1. Load Data
data = load_breast_cancer()
X, y = data.data, data.target

# 2. Define the Search Space
param_grid = {
'n_estimators': [50, 100, 200],
'max_depth': [None, 10, 20],
'min_samples_split': [2, 5, 10]
}

# 3. Initialize GridSearch with 3-Fold Cross-Validation
rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
cv=3, scoring='accuracy', verbose=1)

# 4. Run the search
grid_search.fit(X, y)
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Cross-Validation Score:{grid_search.best_score_:.4f}")
```

### Baseline Model

```python
from sklearn.model_selection import cross_val_score

baseline_model = RandomForestClassifier(random_state=42)
baseline_scores = cross_val_score(baseline_model, X, y, cv=3, scoring='accuracy')

print("Baseline Accuracy:", round(baseline_scores.mean(), 4))
```

### Performance Comparison

```python
print("\nPerformance Comparison:")
print("Baseline Score:", round(baseline_scores.mean(), 4))
print("Tuned Score:", round(grid_search.best_score_, 4))
```

### Total Fits Calculation

```python
total_combinations = (
    len(param_grid['n_estimators']) *
    len(param_grid['max_depth']) *
    len(param_grid['min_samples_split'])
)

total_fits = total_combinations * 3

print("Total Combinations:", total_combinations)
print("Total Model Fits:", total_fits)
```

### Save Model

```python
import joblib

joblib.dump(grid_search.best_estimator_, "best_model.pkl")

print("Model saved successfully!")
```

## Reflect

#### GridSearch tests every combination. If we have 5 parameters with 10 options each, how many models will it train?

If we have 5 parameters with 10 options each, then GridSearchCV will test:
10^5 = 100,000 combinations

If we use 3-fold cross-validation, the total number of models trained will be:
100,000×3=300,000 models

#### Why use RandomizedSearchCV for huge datasets?

GridSearchCV tries every possible combination, which becomes computationally expensive and time-consuming for large datasets or many parameters.
RandomizedSearchCV solves this by:
- Testing only a random subset of combinations
- Reducing training time significantly
- Still finding near-optimal hyperparameters

Therefore, RandomizedSearchCV is preferred for large datasets or very large search spaces because it is much more efficient while maintaining good performance.
